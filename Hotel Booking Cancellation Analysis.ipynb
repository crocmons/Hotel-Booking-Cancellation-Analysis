# Data Analysis and visualizations:

## Analysis and Findings the Hotel Booking Cancellation:

```python
cancelled_percent = df['is_canceled'].value_counts(normalize = True)
cancelled_percent
```

```python
plt.figure(figsize=(5, 4))
bar_colors = ['lightsalmon', 'tab:red']
plt.title('Hotel Reservation Status Count')
plt.bar(['Not Canceled','Canceled'], df['is_canceled'].value_counts(), edgecolor = 'k', width=0.7, color=bar_colors)
plt.show()
```

In here, the company bar graph represents the percentage of hotel reservations that are canceled and those are not canceled. It is obvious that there still a significant number of reservations that have not been canceled. But it is important to notice that there are still 37% of clients are canceled their hotel booking. Which has a significant impact on the hotels potential earnings and we can't ignore it.

```python
# Convert 'is_canceled' values to strings
df['is_canceled'] = df['is_canceled'].astype(str)
```

```python
plt.figure(figsize=(8, 4))
ax1 = sns.countplot(x='hotel', hue='is_canceled', data = df, palette = 'Reds')
legend_labels,_ = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title("Reservation status in different hotels", size=20)
plt.xlabel('Hotel')
plt.ylabel('Number of reservations')
plt.legend(['not canceled', 'canceled'])
plt.show()
```

In comparison to Resort Hotels, City Hotels have more bookings. It's possible that resort hotels are more expensive than those city hotels so people are more prefer to go with city hotels

```python
resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
```

## Resort Hotel Cancelation Rate - 27%
## City Hotel Cancelation Rate - 42%

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20, 8))
plt.title('Average Daily Rate of the City Hotel', fontsize=30)
plt.plot(resort_hotel.index, resort_hotel['adr'], label = 'Resort Hotel', color='indianred')
plt.plot(city_hotel.index, city_hotel['adr'], label = 'City Hotel', color='yellowgreen')
plt.legend(fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(20, 8))
plt.title('Average Daily Rate of the City Hotel and the Resort Hotel', fontsize=30)
plt.scatter(resort_hotel.index, resort_hotel['adr'], marker='^', label = 'Resort Hotel')
plt.scatter(city_hotel.index, city_hotel['adr'], marker='*', label = 'City Hotel')
plt.legend(fontsize=20)
plt.show()
```

The scatter plot or line graph above shows that, on certain days, the average daily rate for a city hotel is less than that of a resort hotel, and on other days, it is even lesser. It goes without saying that weekends and holidays may see a rise in resort hotels. 

```python
df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize=(16, 8))
ax1 = sns.countplot(x='month', hue='is_canceled', data = df, palette = 'Reds')
legend_labels, _ = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title('Reservation Status Per Month', size=20)
plt.xlabel('Month')
plt.ylabel('Number of Reservations')
plt.legend(['Not Canceled', 'Canceled'])
plt.show()
```

In here, I have developed the grouped bar graph to analyze the months with the highest and lowest hotel booking levels according to reservation status date. As can be seen, both the number of confirmed reservations and not canceled are largest in the month of August(08), whereas January(01) is the month with the most canceled reservations.

```python
df['is_canceled'] = df['is_canceled'].astype(int)
```

```python
# Filter data for canceled bookings and group by month, then sum the 'adr' values
grp_canceled_data = df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index()
grp_canceled_data
```

```python
#plot average daily rate for each month only for canceled rate
fig, ax = plt.subplots()
# plt.figure(figsize=(15, 8))
plt.title('ADR Per Month', fontsize = 20)
ax.bar('month', 'adr', data=grp_canceled_data, color='darksalmon')
plt.show();
```

this bar graph demonstrates that cancellations are most common when the prices are highest and are least common when they are lowest. Therefore we can see that the cost of the accommodation is solely responsible for the hotel booking cancellation.

```python
cancelled_adr = canceled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_adr.reset_index(inplace=True)
cancelled_adr.sort_values('reservation_status_date', inplace=True)
```

```python
not_canceled_data = df[df['is_canceled'] == 0]
```

```python
not_cancelled_adr = not_canceled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_adr.reset_index(inplace=True)
not_cancelled_adr.sort_values('reservation_status_date', inplace=True)
```

```python
plt.figure(figsize=(20, 6))
plt.title('Average Daily Rate', fontsize=30)
plt.scatter(not_cancelled_adr['reservation_status_date'], not_cancelled_adr['adr'], label = 'Hotel Booking Not Cancelled')
plt.scatter(cancelled_adr['reservation_status_date'], cancelled_adr['adr'], marker='*', label = 'Hotel Booking Cancelled')
plt.legend(fontsize=17);
```

```python
cancelled_adr = cancelled_adr[(cancelled_adr['reservation_status_date'] > '2016') & (cancelled_adr['reservation_status_date'] < '2017-09')]
not_cancelled_adr = not_cancelled_adr[(not_cancelled_adr['reservation_status_date'] > '2016') & (not_cancelled_adr['reservation_status_date'] < '2017-09')]
```

```python
plt.subplots(figsize=(20, 6))
plt.title('Average Daily Rate', fontsize=30)
plt.scatter(not_cancelled_adr['reservation_status_date'], not_cancelled_adr['adr'], label = 'Hotel Booking Not Cancelled', color='green')
plt.scatter(cancelled_adr['reservation_status_date'], cancelled_adr['adr'],  label = 'Hotel Booking Cancelled', marker='^', color='red')
plt.legend(fontsize=20);
```

## How we gonna Decrease the Hotel Booking Cancelation Rate:

* Cancellation rate rise as the price does. In order to prevent cancellations of reservations, hotels could work on their pricing strategies and try to lower the rates for specific hotels based on locations and countries. They can also provide some discounts to the consumers.

* As the ratio of the cancellation and not cancellation of the resort hotel is higher in the resort hotel than the city hotels. So the hotels should provide a reasonable discount on the room prices on weekends or on holidays.

* In the month of January, hotels can start campaigns or marketing with a reasonable amount to increase their revenue as the cancellation is the highest in this month.

* They can also increase the quality of their hotels and their services mainly in Portugal to reduce the cancellation rate.
